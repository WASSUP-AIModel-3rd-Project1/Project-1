```python
import pandas as pd
import numpy as np
import os
```

```python
DATASET_DIR = ''
dir_name = './'
file_name = 'BigCitiesHealth.csv'
```

```python
file_path = os.path.join(DATASET_DIR,dir_name,file_name)
df = pd.read_csv(file_path)
```

```python
df["metric_label"] = df["metric_subcat_label"]+' | '+df["metric_item_label"]

cond0 = df['geo_label_city'] == 'U.S. Total'

df_US = df.loc[cond0].copy()
df_city = df.loc[~cond0].copy()
```

```python
df_baseline = df_city[(df_city["strata_race_label"] == "All") & (df_city["strata_sex_label"] == "Both")]
```

```python
table = pd.pivot_table(df_baseline, index=["date_label", "geo_label_city"], values='value', columns=['metric_label'])
```

```python
table = table.reset_index(1)
```

```python
# stratify by year
years = df["date_label"].unique()
years.sort()
years
```

```python
# ì—°ë„ë³„ ì¸ë±ì‹± ìž˜ ëë‚˜ ë³´ê¸° ìœ„í•´ 2010ë…„ ë°ì´í„° í™•ì¸
table.loc[2010,:]
```

```python
by_year = []
for year in years:
    by_year.append(table.loc[year,:])
```

```python
# ì—°ë„ë³„ë¡œ ëžœë¤í•˜ê²Œ 20%ì”© í…ŒìŠ¤íŠ¸ë¡œ ë¶„ë¦¬

from sklearn.model_selection import train_test_split

train = []
test = []
for strata in by_year:
    curr_train, curr_test = train_test_split(strata, test_size=0.2, random_state=42)
    train.append(curr_train)
    test.append(curr_test)

train = pd.concat(train)
test = pd.concat(test)
```

```python
test_scaled = (test.drop('geo_label_city', axis=1) - train.drop('geo_label_city', axis=1).min()) / (train.drop('geo_label_city', axis=1).max() - train.drop('geo_label_city', axis=1).min())
test_scaled.insert(0, "geo_label_city", test.loc[:,"geo_label_city"])
```
